```python
import pandas as pd
url='https://github.com/marianalarra/ChallengeTelecom2/blob/main/datos_tratados.csv'
```

```python
import requests

url = 'https://raw.githubusercontent.com/marianalarra/ChallengeTelecom2/main/datos_tratados.csv'
response = requests.get(url)
response.raise_for_status() # Raise an exception for bad status codes
with open('datos_tratados.csv', 'wb') as f:
    f.write(response.content)

df = pd.read_csv('datos_tratados.csv')
display(df.head())
```
