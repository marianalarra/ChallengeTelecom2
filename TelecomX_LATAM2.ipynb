###1-Extraccion del archivo tratado

```python
import pandas as pd
url='https://github.com/marianalarra/ChallengeTelecom2/blob/main/datos_tratados.csv'
```

```python
import requests

url = 'https://raw.githubusercontent.com/marianalarra/ChallengeTelecom2/main/datos_tratados.csv'
response = requests.get(url)
response.raise_for_status() # Raise an exception for bad status codes
with open('datos_tratados.csv', 'wb') as f:
    f.write(response.content)

df = pd.read_csv('datos_tratados.csv')
display(df.head())
```

###2- Transformacion de datos

####1) Eliminación de columnas irrelevantes

```python
df= df.drop(columns=['customerID'])
```

```python
df.sample(5)
```

####2) Variables explicativas y variable de respuesta

```python
X= df.drop(columns=['Churn'])
y= df['Churn']
```

####3) Transformando las variables explicativas

##### Transformacion de columnas yes/no

```python
df['PaperlessBilling'].unique()
```

```python
df.info()
```

```python
cols = ['Partner','Dependents', 'PhoneService', 'PaperlessBilling']
for col in cols:
    df[col] = df[col].str.strip().str.lower().map({"yes": 1, "no": 0})
```

```python
df.sample(5)
```

#####Encoding

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
columnas= X.columns
columnas
```

```python
one_hot = make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore"),
     ['gender','InternetService', 'Contract', 'PaymentMethod']),
    remainder='passthrough',
    sparse_threshold=0
)
```

```python
X_encoded = one_hot.fit_transform(X)
```

```python
X_encoded_df= pd.DataFrame(X, columns = one_hot.get_feature_names_out(columnas))
```
